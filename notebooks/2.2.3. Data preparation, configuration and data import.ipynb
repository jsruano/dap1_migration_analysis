#### 2.2.3.1 **World Bank Indicators**

For the World Bank data, I first refined the list of relevant countries by excluding categories like regions and areas from the Excel file. I then imported this refined data into the **wb_countries** table in **Navicat**, resulting in a table with 216 records, each representing a country with its relevant information. This table served as a reference for comparing with other indicator tables imported separately. Subsequently, the data from the four separate indicator tables were consolidated into a single table, **wb_values**. SQL queries were used to verify and compare record counts across tables to ensure data consistency, and joins were performed to integrate the tables into the final data model.

Here is an example of a join between **wb_countries**, with the refined list of countries, and **____wb_gdppc**, the imported table for one of the indicators:

```python
from sqlalchemy import create_engine
import pandas as pd

# Configure connection chain
engine = create_engine('mysql+pymysql://root:datapoet@localhost/dap1')

# Connect and execute query
query = '''SELECT c.country_code, g.country_code AS country_code_gdp
FROM wb_countries AS c
LEFT JOIN ____wb_gdppc AS g 
     ON c.country_code = g.country_code
WHERE g.country_code IS NULL;'''

df = pd.read_sql(query, engine)

# Show results
df.head()
```

Here, we see an example of a join between **wb_countries**, which contains the predefined list of countries, and **____wb_gdppc**, one of the imported indicator tables. By comparing the **country_code** from both tables using a LEFT JOIN, it is evident that there are no discrepancies between them. The country code serves as the common key for both tables. Subsequently, after verification, all four tables were merged into the parent table **wb_values** using the following query:

```python
from sqlalchemy import create_engine
import pandas as pd
from IPython.display import display, HTML

# Configure the connection string
engine = create_engine('mysql+pymysql://root:datapoet@localhost/dap1')

# Query and display the wb_countries table
query_countries = '''
SELECT * FROM wb_countries
LIMIT 15;
'''
df_countries = pd.read_sql(query_countries, engine)

# Display the DataFrame as an HTML table
print("wb_countries Table:")
display(HTML(df_countries.to_html()))

# Query and display the wb_values table
query_values = '''
SELECT * FROM wb_values
LIMIT 15;
'''
df_values = pd.read_sql(query_values, engine)

# Display the DataFrame as an HTML table
print("wb_values Table:")
display(HTML(df_values.to_html()))

# Query and display the wb_indicators table
query_indicators = '''
SELECT * FROM wb_indicators
LIMIT 15;
'''
df_indicators = pd.read_sql(query_indicators, engine)

# Display the DataFrame as an HTML table
print("wb_indicators Table:")
display(HTML(df_indicators.to_html()))
```
