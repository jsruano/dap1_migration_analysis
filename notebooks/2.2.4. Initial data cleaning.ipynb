The data cleaning process in **wb_countries** involved the following queries, creating a new view called **un_locations_view1**. Through this query:
*SELECT id, name FROM un_locations WHERE BINARY name = UPPER(name)* only the uppercase names are obtained in order to transform them:

```python
%%sql
WITH un_locations_cleaning1 AS (
SELECT id, TRIM(BOTH '*' FROM name) AS name
FROM un_locations
),

un_locations_cleaning2 AS (
SELECT id, REPLACE(REPLACE(REPLACE(TRIM(name), '\n', ''), '\r', ''), '\t', '') AS name
FROM un_locations_cleaning1
),

un_locations_cleaning3 AS (
SELECT id,
  CASE
    WHEN name = 'WORLD' THEN 'World'
    WHEN name = 'AFRICA' THEN 'Africa'
    WHEN name = 'LATIN AMERICA AND THE CARIBBEAN' THEN 'Latin America and the Caribbean'
    WHEN name = 'NORTHERN AMERICA' THEN 'Northern America'
    WHEN name = 'EUROPE' THEN 'Europe'
    WHEN name = 'OCEANIA' THEN 'Oceania'
    WHEN name = 'ASIA' THEN 'Asia'
  ELSE name
  END AS name
FROM un_locations_cleaning2
)

SELECT id, REPLACE (name, 'And', 'and') AS name
FROM un_locations_cleaning3;
```
